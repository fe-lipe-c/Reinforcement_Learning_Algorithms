```python
import gym
import pandas as pd
```

```python
all_envs = gym.envs.registry.all()
env_ids = [env_spec.id for env_spec in all_envs]
print('Total number of enviroments: ',len(env_ids),'\n')
print('Cartpole enviroments: ',sorted([word for word in env_ids if word.startswith('m')]))
```

```python
enviroments = pd.DataFrame()

action_space = []
observation_space = []
error = []

for ind,game in enumerate(env_ids):
    
    try:
        if game.startswith('Defender'):
            action_space.append('pass')
            observation_space.append('pass')
            continue
        env = gym.make(game)
        action_space.append(env.action_space)
        observation_space.append(env.observation_space)
        env.close
    except:
        error.append(game)
        action_space.append('error')
        observation_space.append('error')
        pass
    
enviroments['Game'] = env_ids
enviroments['Obs_Space'] = observation_space
enviroments['Action_Space'] = action_space

```
